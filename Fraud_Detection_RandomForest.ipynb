```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
in_data=pd.read_csv('C:\Users\AG28744\Desktop\Projects\Misc\DS_Datasets\SVM_healthcare-provider-fraud-detection-analysis\SVM_Train.csv')
in_data.head()
```

```python
in_data.shape
```

```python
train_prdctrs=in_data[in_data.columns[1:32]]
train_prdctrs.head()
```

```python
from sklearn import preprocessing
```

```python
train_prdctrs_std=preprocessing.scale(train_prdctrs)
train_prdctrs_std=pd.DataFrame(train_prdctrs_std)
train_prdctrs_std.head()
```

```python
train_op=in_data[in_data.columns[32]]
train_op.head()
```

```python
train_op.shape
```

```python
from sklearn.ensemble import RandomForestRegressor    #model building
```

```python
regressor = RandomForestRegressor(n_estimators=20, random_state=0)
```

```python
regressor.fit(train_prdctrs_std,train_op)
```

```python
predtd_fraud1=regressor.predict(train_prdctrs_std)
predtd_fraud1=pd.DataFrame(predtd_fraud1)
in_data['predtd_fraud1']=predtd_fraud1
in_data.head()
```

```python
in_data["predtd_value"]=0
```

```python
in_data.loc[in_data.predtd_fraud1 > 0.5, "predtd_value"] = 1
in_data.head()
```

```python
test_data2=test_data[test_data.columns[1:32]]
test_data2.head()
```
