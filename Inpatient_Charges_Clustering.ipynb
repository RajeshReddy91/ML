```python
from sklearn import preprocessing
data_std=preprocessing.scale(n_data)
```

```python
data_std=pd.DataFrame(data_std)
```

```python
data_std.head()
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=5)
model.fit(data_std)
```

```python
model.labels_
```

```python
#type(model_data)
```

```python
model_data=pd.DataFrame(model.labels_)
model_data.head()
```

```python
n_data['cluster_no']=model_data
n_data.head()
```

```python
provider_data=n_data.iloc[:,[6,0,1,2,3,4,5]]
provider_data.head()
```

```python
provider_summary=provider_data.iloc[:,[3,4,5,6]].groupby(provider_data.cluster_no).mean()
provider_summary
```

```python
# from above details we can say that providers in cluster 4 provide good service
```

```python
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(k_data)
    k_data["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
model_2=KMeans(n_clusters=4)
model_2.fit(data_std)
model2_data=pd.DataFrame(model_2.labels_)
model2_data.head()
```

```python
k_data['cluster_no']=model2_data
provider_data2=k_data.iloc[:,[6,0,1,2,3,4,5]]
provider_summary2=provider_data2.iloc[:,[3,4,5,6]].groupby(provider_data2.cluster_no).mean()
provider_summary2
```
